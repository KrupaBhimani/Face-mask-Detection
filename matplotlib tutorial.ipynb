```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
plt.plot([1,2,3],[4,5,1]) #providing x axis and yaxis values
```

```python
x = [5,8,10]
y = [12,18,6]

plt.plot(x,y)

#adding tilte and labels
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

```

```python
style.use('ggplot')

x=[3,8,7]
y=[10,5,16]

x2=[6,9,11]
y2=[6,15,7]

plt.plot(x,y,'g',label='Lineone',linewidth=5)
plt.plot(x2,y2,'c',label='Linetwo',linewidth=5)

#adding tilte and labels
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()

plt.grid(True,color='k')
```

```python
#bar graph
plt.bar([1,3,5,7,9],[5,2,7,8,2],label="ExampleOne")

plt.bar([2,4,6,8,10],[8,6,2,5,6],label="ExampleTwo",color='g')

plt.legend()
plt.ylabel('bar number')
plt.xlabel('bar height')

plt.title('SGP 3')

```
